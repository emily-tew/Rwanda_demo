```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_excel('/Users/emilytew/Documents/rwanda/Avocado.xlsx', 'Sheet1')
display(df)
```

```python
df.loc['Total'] = pd.Series(df['AveragePrice'].sum(), index = ['AveragePrice'])
display(df)
```

```python
# gives sums of the columns
df2 = df.sum(axis=0)
display(df2)
```

```python
# adds those sums onto the bottom of the dataframe

df.loc['sum'] = df[:-1].sum() #This [:-1] means it does not add the 'Total' column we just created above
display(df)
```
